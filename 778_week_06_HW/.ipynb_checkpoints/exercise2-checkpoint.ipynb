Uses the NASA Images and Video Library API and `pandas` to display images based on user-provided search term and limiter.

```python
import pandas as pd

term = input("Enter a search term.")

page = 1 # could be used in a loop to access all returned results across multiple pages, but who has the time?

response = pd.read_json(f"https://images-api.nasa.gov/search?q={term}&page={page}&media_type=image")

frame = pd.json_normalize(response['collection']['items'],record_path='data')

print(f"Found {len(frame)} results based on search term '{term}'. Run the next cell to continue.")
```

```python
years = frame['date_created'].str[:4]
year = input(f"Image results come from years {years.min()}-{years.max()}. Enter a year to narrow your search.")

filtered = frame.loc[frame['date_created'].str.contains(year)]

print(f"Found {len(filtered)} results based on search term '{term}' and year {year}. Run the next cell to continue.")
```

```python
if len(filtered) > 1:
    index = input(f"Choose an index from 0 to {len(filtered)-1}.")
else:
    index = 0

nasa_id = filtered.iloc[index]['nasa_id']
   
from urllib.request import urlopen
from PIL import Image
import json
from IPython.display import display
 
image_url = json.load(urlopen(f"https://images-assets.nasa.gov/image/{nasa_id}/collection.json"))[0]
image = Image.open(urlopen(image_url))
display(image)
```
