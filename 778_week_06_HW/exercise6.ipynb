Uses [NYS's Incarcerated Individuals Under Custody](https://data.ny.gov/Public-Safety/Incarcerated-Individuals-Under-Custody-Beginning-2/55zc-sp6m/about_data) dataset to look up the age and most serious crime of every person incarcerated in a user-provided county.

```python
import json
from urllib.request import urlopen
from urllib.parse import quote

query = input("Choose a county in New York State.")

county = quote(query).upper()

api_endpoint = f"https://data.ny.gov/resource/55zc-sp6m.json?county_of_indictment={county}&snapshot_year=2023"

temp_results = json.load(urlopen(api_endpoint))
results = temp_results.copy()
offset = 1000

while len(temp_results) > 999:
    temp_results = json.load(urlopen(api_endpoint + "&$offset=" + str(offset)))
    results.extend(temp_results)
    offset += 1000

if len(results) == 0:
    print("No inmates found for " + query + " County.")
else:
    print("Found the following inmates for " + query + " County:")
    for index, inmate in enumerate(results):
        print(str(index+1) + ". Age " + str(inmate['current_age']) + ", Most Serious Crime: " + inmate['most_serious_crime'])
    print("Found " + str(len(results)) + " inmates total.")
```
